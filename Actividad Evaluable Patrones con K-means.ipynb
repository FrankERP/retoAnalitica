```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances_argmin_min
 
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
plt.rcParams['figure.figsize'] = (16, 9)
plt.style.use('ggplot')
```

```python
dataframe = pd.read_csv('avocado.csv')
dataframe.head()
```

```python
dataframe.describe()
```

```python
print(dataframe.groupby('year').size())
```

```python
dataframe.drop(['year'],1).hist()
plt.show()
```

```python
sb.pairplot(dataframe.dropna(), hue='year',height=4,vars=["AveragePrice","Total Volume","Total Bags"],kind='scatter')
```

```python
X = np.array(dataframe[["AveragePrice","Total Volume","Total Bags"]])
y = np.array(dataframe['year'])

X.shape
```

```python
fig = plt.figure()
ax = Axes3D(fig)
colores=['blue','red','green','yellow','cyan','orange']
#NOTA: asignamos la posición cero del array repetida pues las categorTas comienzan en id 1. 
asignar=[]
for row in y:
    asignar.append(colores[row-2015])
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=asignar,s=60)
```

```python
Nc = range(1, 20)
kmeans = [KMeans(n_clusters=i) for i in Nc]
kmeans
score = [kmeans[i].fit(X).score(X) for i in range(len(kmeans))]
score
plt.plot(Nc,score)
plt.xlabel('Number of Clusters')
plt.ylabel('Score')
plt.title('Elbow Curve')
plt.show()
```

```python
kmeans = KMeans(n_clusters=5).fit(X)
centroids = kmeans.cluster_centers_
print(centroids)
```

```python
# Predicting the clusters
labels = kmeans.predict(X)
# Getting the cluster centers
C = kmeans.cluster_centers_
colores=['red','green','blue','cyan','yellow']
asignar=[]
for row in labels:
    asignar.append(colores[row])

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=asignar,s=60)
ax.scatter(C[:, 0], C[:, 1], C[:, 2], marker='*', c=colores, s=1000)
```

```python
# Getting the values and plotting it
f1 = dataframe['AveragePrice'].values
f2 = dataframe['Total Volume'].values

plt.scatter(f1, f2, c=asignar, s=70)
plt.scatter(C[:, 0], C[:, 1], marker='*', c=colores, s=1000)
plt.show()
```

```python
# Getting the values and plotting it
f1 = dataframe['AveragePrice'].values
f2 = dataframe['Total Bags'].values

plt.scatter(f1, f2, c=asignar, s=70)
plt.scatter(C[:, 0], C[:, 2], marker='*', c=colores, s=1000)
plt.show()
```

```python
f1 = dataframe['Total Bags'].values
f2 = dataframe['Total Volume'].values

plt.scatter(f1, f2, c=asignar, s=70)
plt.scatter(C[:, 1], C[:, 2], marker='*', c=colores, s=1000)
plt.show()
```
